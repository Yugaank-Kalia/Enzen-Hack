# Python script for handling data
#### Importing Dependencies like **pandas, numpy**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.patches as patch
from IPython.display import Image
```

#### Creating dataframes using pandas

```python
load = pd.read_csv('load.csv')
load.columns = ['12am', '1am', '2am', '3am', '4am', '5am', '6am', '7am', '8am', '9am', '10am', '11am', '12pm', '1pm', '2pm', '3pm', '4pm', '5pm', '6pm', '7pm', '8pm', '9pm', '10pm', '11pm']
load = load.apply(lambda row: row/100000)
print(load.head(5))
```

```python
solar = pd.read_csv('solar.csv')
solar.columns = ['12am', '1am', '2am', '3am', '4am', '5am', '6am', '7am', '8am', '9am', '10am', '11am', '12pm', '1pm', '2pm', '3pm', '4pm', '5pm', '6pm', '7pm', '8pm', '9pm', '10pm', '11pm']
solar = solar.apply(lambda row: row*1000)
print(solar.head(10))
```

### Plot of load and solar output

```python
x_axis = load.columns

y1 = load.head(1)

y2 = solar.head(1)

plt.plot(y1, color="red")
plt.plot(y2, color="blue")

plt.title("Production of Solar Energy")

plt.xticks(np.arange(len(x_axis)), x_axis, rotation=90)

x_label = plt.xlabel("Time")
y_label = plt.ylabel("Power Generated")

red = patch.Patch(color="red", label="Load")
blue = patch.Patch(color="blue", label="Solar")

# plt.legend(handles = [red, blue])

# plt.show()
```
